## Import necessary modules

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise 2.1

```python
sum_ = 0
# Loop through the 10,000 vocabulary terms and compute the perplexity
for i in range(1,10001):
    sum_ += (0.1 / (i)) * math.log(0.1 / (i))
sum_ = math.exp(-sum_)
print("Perplexity is", sum_)
```

## Exercise 2.2

```python
# Perplexity for varying c
v = 10000
c = np.arange(0.01,0.51,0.01)
list_sum = []
for k in c:
    _sum = 0
    for i in range(1,v+1):
        _sum += (k / (i)) * math.log(k / (i))
    _sum = math.exp(-_sum)
    list_sum.append(_sum)
plt.plot(c,list_sum);
plt.xlabel('Value of c')
plt.ylabel('Perplexity')
plt.tight_layout()
plt.title('Perplexity for varying c with vocabulary size '+str(v))
plt.show()
```

```python
# Perplexity for varying v
c = 0.1
v = list(np.arange(100, 1000000, 4000))
#print(v[-1])
v.append(1000000)
list_sum = []
for k in v:
    _sum = 0
    for i in range(1,k+1):
        _sum += (c / (i)) * math.log(c / (i))
    _sum = math.exp(-_sum)
    list_sum.append(_sum)
plt.plot(v,list_sum);
plt.xlabel('Value of vocabulary size')
plt.ylabel('Perplexity')
plt.tight_layout()
plt.title('Perplexity for varying vocabulary size with c '+str(c))
plt.show()
```

## Problem 3.2

```python
# Compute the perplexity
probs = [2/11,2/11,2/11,1/11,1/11,1/11,1/11,1/11]
sum_ = 0
for i in probs:
    sum_ += (i * math.log(i))
prplx = math.exp(-sum_)
print('Perplexity is',prplx)
```
